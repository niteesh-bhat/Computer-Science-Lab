```python
class Graph:
    def __init__(self, graph, hueristicNodeList, source):
        self.graph=graph
        self.H=hueristicNodeList
        self.parent={}
        self.status={}
        self.solution={}
        self.source=source
    
    def applyAOStart(self):
        self.aoStar(self.source, False)
        
    def getNeighbors(self, v):
        return self.graph.get(v, '')
        
    def getStatus(self, v):
        return self.status.get(v, 0)
    
    def setStatus(self, v, val):
        self.status[v]=val
    
    def getHueristicNodeValue(self, n):
        return self.H.get(n, 0)
    
    def setHueristicNodeValue(self, n, val):
        self.H[n]=val
        
    def printSolution(self):
        print("For graph solution, Traverse the graph from start Node: ", self.source)
        
        print("Hueristic values: ", self.H)
        print("Solution Graph: ", self.solution)
        
    def computeMinCostChildNodes(self, v):
        minCost=0
        costToChildNodeList={}
        costToChildNodeList[minCost]=[]
        flag=True
        
        for nodeInfo in self.getNeighbors(v):
            cost=0
            nodeList=[]
            
            for c, wt in nodeInfo:
                cost+=self.getHueristicNodeValue(c) +wt
                nodeList.append(c)
            
            if flag==True:
                minCost=cost
                costToChildNodeList[minCost]=nodeList
                flag=False
            else:
                if minCost>cost:
                    minCost=cost
                    costToChildNodeList[minCost]=nodeList
                    flag=False
        
        return minCost, costToChildNodeList[minCost]
    
    def aoStar(self, v, isBacktracking):
        print("Processing node: ", v)
        
        if self.getStatus(v) >= 0:
            minCost, childNodeList = self.computeMinCostChildNodes(v)
            print(minCost, childNodeList)
            
            self.setHueristicNodeValue(v, minCost)
            self.setStatus(v, len(childNodeList))
            solved=True
            
            for childNode in childNodeList:
                self.parent[childNode]=v
                if self.getStatus(childNode)!=-1:
                    solved=solved and False
            
            if solved==True:
                self.setStatus(v, -1)
                self.solution[v]=childNodeList
                
            if v!=self.source:
                self.aoStar(self.parent[v], True)
            
            if isBacktracking == False:
                for childNode in childNodeList:
                    self.setStatus(childNode, 0)
                    self.aoStar(childNode, False)
```

```python
H1={
    'A': 1,
    'B': 6,
    'C': 2,
    'D': 12,
    'E': 2,
    'F': 1,
    'G': 5,
    'H': 7,
    'I': 7,
    'J': 1,
}

graph1={
    'A': [[('B', 1), ('C', 1)], [('D', 1)]],
    'B': [[('G', 1)], [('H', 1)]],
    'C': [[('J', 1)]],
    'D': [[('E', 1), ('F', 1)]],
    'G': [[('I', 1)]]
}

G1=Graph(graph1, H1, 'A')
G1.applyAOStart()
G1.printSolution()
```
