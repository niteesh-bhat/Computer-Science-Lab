```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plot_vector2D(label, vector, color='r', origin=(0,0)):
    origin=np.array(origin)
    vector=np.array(vector)
    
    plt.quiver(*origin, *vector, color=color, angles='xy', scale_units='xy', scale=1, width=0.005)
    plt.text(*(vector+origin+0.1), label)
```

```python
A=np.array([[3, 1],[1, 3]])
B=np.array([2,1])
X=np.linalg.inv(A).dot(B)
print(X)
```

```python
# Uniform mesh grid and basis vector plot
z=4
plt.xlim([-z, z])
plt.ylim([-z, z])

x_cords=np.linspace(-z, z, 2*z+1)
y_cords=np.linspace(-z, z, 2*z+1)
grid=np.array(np.meshgrid(x_cords, y_cords)).reshape(2, (2*z+1)**2)
```

```python
plt.scatter(*grid, s=2,alpha=0.5)
plt.scatter(*B, s=5, color='c')
plt.text(*(B+0.1), 'c')

plot_vector2D('basis 1',[1,0], color='r')
plot_vector2D('basis 2',[0,1], color='b')
plot_vector2D('x', X, color='g')
```

```python
# Transformed grid
z=4
plt.xlim([-z, z])
plt.ylim([-z, z])

plt.scatter(*(np.dot(A, grid)), s=2,alpha=0.5, color='r')
plt.scatter(*B, s=5, color='c')
plt.text(*(B-0.25), 'c')

plot_vector2D('Tbasis 1',np.dot(A, np.array([1,0]).T), color='r')
plot_vector2D('Tbasis 2',np.dot(A, np.array([0,1]).T), color='b')
plot_vector2D('Tx', np.dot(A, X.T), color='g')
```
